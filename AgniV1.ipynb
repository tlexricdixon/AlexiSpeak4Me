```python
# Cell 2: High-Power Agni-Gemma-12B Setup
from unsloth import FastLanguageModel
import torch

max_seq_length = 4096 # High-end context for complex records
dtype = None # Auto-detect (A100 will use bfloat16)
load_in_4bit = True # Keeps VRAM usage comfortable

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/gemma-3-12b-it",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)

# Deep "Brain Surgery" for the 12B model
model = FastLanguageModel.get_peft_model(
    model,
    r = 64,  # Increased to 64: Giving the 12B model a massive "learning buffer"
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 64,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
    random_state = 3407,
)
```

```python
import json
import random
from datasets import Dataset

def generate_business_master_dataset(count=5000):
    instructions = []

    for i in range(count):
        choice = random.random()

        # --- TASK 1: MATH & TAX (40%) ---
        if choice < 0.40:
            price = random.randint(50, 2000)
            qty = random.randint(1, 5)
            subtotal = price * qty
            tax = subtotal * 0.10
            total = subtotal + tax

            instructions.append({
                "instruction": "Calculate the total for this business expense including 10% tax.",
                "input": f"Item: Office Hardware, Price: ${price}, Quantity: {qty}",
                "output": f"Steps: {qty}x Hardware @ ${price} = ${subtotal}. Tax (10%): ${tax:.2f}. Total: ${total:.2f}."
            })

        # --- TASK 2: SALES FORECASTING (30%) ---
        elif choice < 0.70:
            current = random.randint(10000, 50000)
            growth = random.choice([3, 5, 8, 12])
            increase = current * (growth / 100)

            instructions.append({
                "instruction": f"Project the Q3 outlook with a {growth}% growth target.",
                "input": f"Current Revenue: ${current}",
                "output": f"Analysis: A {growth}% increase adds ${increase:.2f}. Predicted Q3 Revenue: ${current + increase:.2f}."
            })

        # --- TASK 3: DB RECORD EXTRACTION (30%) ---
        else:
            ref_id = f"REF-{random.randint(100, 999)}"
            client = random.choice(["TechNova", "GlobalLogistics", "ApexSolutions"])
            status = random.choice(["PAID", "PENDING", "OVERDUE"])

            instructions.append({
                "instruction": "Extract the database record into a readable summary.",
                "input": f"ID:{ref_id} | CLNT:{client} | STTS:{status} | RMK:N/A",
                "output": f"Summary: Record {ref_id} for client {client} is currently marked as {status}."
            })

    # Convert to HuggingFace Dataset format
    dataset = Dataset.from_list(instructions)
    return dataset

# 1. Generate the data
dataset = generate_business_master_dataset(5000)

# 2. Apply Gemma 3 Formatting (Crucial step!)
from unsloth.chat_templates import get_chat_template

tokenizer = get_chat_template(
    tokenizer,
    chat_template = "gemma-3", # Specifically triggers Gemma 3 tokens
)

def formatting_prompts_func(examples):
    instructions = examples["instruction"]
    inputs       = examples["input"]
    outputs      = examples["output"]
    texts = []
    for instruction, input, output in zip(instructions, inputs, outputs):
        # We wrap the user prompt and the model response in Gemma's specific 'turns'
        convo = [
            {"role": "user", "content": f"{instruction}\n\n{input}"},
            {"role": "assistant", "content": output},
        ]
        text = tokenizer.apply_chat_template(convo, tokenize = False, add_generation_prompt = False)
        texts.append(text)
    return { "text" : texts, }

dataset = dataset.map(formatting_prompts_func, batched = True)

print(f"Success! Created {len(dataset)} Gemma-formatted examples.")
```

```python
# Cell 3: Training Agni-Gemma-12B
from trl import SFTTrainer
from transformers import TrainingArguments

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    args = TrainingArguments(
        per_device_train_batch_size = 4, # A100 can handle this easily
        gradient_accumulation_steps = 4,
        warmup_steps = 20,
        max_steps = 250,                # Extra steps to saturate the 12B brain
        learning_rate = 1e-4,           # Slightly lower LR for larger model stability
        bf16 = True,                    # Pro Plan A100 special sauce
        logging_steps = 1,
        optim = "adamw_8bit",
        output_dir = "Agni-Gemma-12B-Checkpoints",
    ),
)

# Apply the Gemma 3 specific response masking
from unsloth.chat_templates import train_on_responses_only
trainer = train_on_responses_only(
    trainer,
    instruction_part = "<start_of_turn>user\n",
    response_part = "<start_of_turn>model\n",
)

trainer.train()
```

```python
# Cell 4: Exporting the Masterpiece
model.save_pretrained("Agni-Gemma-12B")
tokenizer.save_pretrained("Agni-Gemma-12B")

# Optional: Save as GGUF for your Mac/Ollama immediately
model.save_pretrained_gguf("Agni-Gemma-12B-GGUF", tokenizer, quantization_method = "q4_k_m")
```

```python
# 1. Enable native 2x faster inference
FastLanguageModel.for_inference(model)

# 2. Define a test prompt (Try one of the business tasks we trained on)
messages = [
    {"role": "user", "content": "Project the Q3 outlook with a 12% growth target.\n\nCurrent Revenue: $45,000"}
]

# 3. Apply the Gemma 3 template
inputs = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt = True, # Crucial for instructing the model to reply
    tokenize = True,
    return_tensors = "pt",
).to("cuda")

# 4. Stream the output token by token (so you don't have to wait!)
from transformers import TextStreamer
text_streamer = TextStreamer(tokenizer, skip_prompt = True)

_ = model.generate(
    input_ids = inputs,
    streamer = text_streamer,
    max_new_tokens = 128,
    temperature = 1.0, # Recommended for Gemma 3
    top_p = 0.95,
)
```
